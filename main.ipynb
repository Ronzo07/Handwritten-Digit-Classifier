```python
# Importing the data loader and model functions
from data_loader import load_data
from neural_network import create_model

# Load the MNIST dataset containing handwritten digit images and labels and preprocess it
(x_train, y_train), (x_test, y_test) = load_data()

# Create the feed-forward neural network model
model = create_model()

# Train the model with the training dataset
# - epochs=5: Model goes through the training data 5 times
# - validation_data: Evaluate performance on test data after each epoch to monitor learning progress
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Evaluate the trained model on the test dataset to check its generalization performance
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_accuracy:.4f}") 
```
